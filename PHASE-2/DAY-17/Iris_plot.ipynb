```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Style
sns.set(style="whitegrid")

# Load Dataset
iris = sns.load_dataset("iris")

# Display first 5 rows
print(iris.head())

# Dataset information
print(iris.info())

# Statistical Summary
print(iris.describe())

# Check missing values
print(iris.isnull().sum())
```

```python
plt.figure(figsize=(10,5))

plt.plot(iris.index, iris["sepal_length"],
         color="blue",
         label="Sepal Length")

plt.title("Line Plot of Sepal Length")
plt.xlabel("Observation Index")
plt.ylabel("Sepal Length (cm)")
plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.barplot(x="species",
            y="sepal_length",
            data=iris,
            palette="Set2")

plt.title("Average Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Average Sepal Length")

plt.show()
```

```python
plt.figure(figsize=(7,5))

plt.hist(iris["petal_length"],
         bins=15,
         color="skyblue",
         edgecolor="black")

plt.title("Histogram of Petal Length")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.scatterplot(x="sepal_length",
                y="petal_length",
                hue="species",
                data=iris,
                s=80)

plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.boxplot(x="species",
            y="sepal_width",
            data=iris,
            palette="Pastel1")

plt.title("Box Plot of Sepal Width")
plt.xlabel("Species")
plt.ylabel("Sepal Width")

plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.countplot(x="species",
              data=iris,
              palette="viridis")

plt.title("Count of Iris Species")
plt.xlabel("Species")
plt.ylabel("Count")

plt.show()
```
