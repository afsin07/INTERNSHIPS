# Day 18 – Practical Assignment: Machine Learning Dataset Research


## Dataset 1: Pima Indians Diabetes Dataset

**Dataset Source:** https://www.kaggle.com/datasets/uciml/pima-indians-diabetes-database

**Problem Statement:** Predict whether a patient has diabetes based on medical measurements.

- **Machine Learning Problem Type:** Classification
- **Number of Features:** 8 input features (9 columns including target)
- **Target Variable:** Outcome

**Reason for Selecting:**
- Healthcare-related dataset
- Beginner-friendly
- Suitable for comparing multiple classification algorithms



## Dataset 2: Wine Quality Dataset

**Dataset Source:** https://archive.ics.uci.edu/dataset/186/wine+quality

**Problem Statement:** Predict wine quality using physicochemical properties.

- **Machine Learning Problem Type:** Classification / Regression
- **Number of Features:** 11 input features (12 columns including target)
- **Target Variable:** Quality

**Reason for Selecting:**
- Real-world industrial dataset
- Good for feature analysis
- Supports both regression and classification



## Dataset 3: Student Performance Dataset

**Dataset Source:** https://archive.ics.uci.edu/dataset/320/student+performance

**Problem Statement:** Predict students' final grades using demographic and academic information.

- **Machine Learning Problem Type:** Classification / Regression
- **Number of Features:** 32 input features (33 columns including target)
- **Target Variable:** G3 (Final Grade)

**Reason for Selecting:**
- Educational dataset
- Rich feature set
- Suitable for predictive modeling



# Final Selected Dataset

## Pima Indians Diabetes Dataset

### Justification
The Pima Indians Diabetes Dataset was selected because it is healthcare-related, well-structured, and widely used for binary classification. It allows easy implementation and comparison of algorithms such as Logistic Regression, KNN, Decision Tree, Random Forest, and SVM.

### Objective

- Select the best-performing model

- Logistic Regression
- K-Nearest Neighbors (KNN)
- Decision Tree
- Random Forest
