# MediPredict — AI Health Risk Assessment Platform

## Project Overview

MediPredict is a machine learning-based health risk assessment platform designed to predict whether a patient is at risk of heart disease using medical and lifestyle information. The project aims to support early disease detection and assist healthcare professionals in decision-making.

---

## Problem Statement

Develop an AI-powered health risk assessment platform capable of predicting heart disease risk based on patient health records.

---

## Machine Learning Problem Type

**Classification**

---

## Project Objectives

- Understand the healthcare dataset.
- Explore and analyze patient data.
- Build machine learning classification models.
- Compare model performance.
- Predict health risk accurately.

---

## Dataset Information

- **Dataset Name:** Heart Disease Prediction Dataset
- **Source:** Kaggle
- **Target Variable:** HeartDisease
- **Number of Records:** 918
- **Number of Features:** 12

---

## Dataset Features

- Age
- Sex
- ChestPainType
- RestingBP
- Cholesterol
- FastingBS
- RestingECG
- MaxHR
- ExerciseAngina
- Oldpeak
- ST_Slope
- HeartDisease (Target)
